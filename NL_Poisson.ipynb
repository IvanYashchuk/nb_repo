```python
from IPython.display import HTML
from dolfin import *
#import dijitso
#dijitso.set_log_level('debug')
parameters["form_compiler"]["representation"] = "uflacs"
import matplotlib.pyplot as plt
%matplotlib inline
!cd ~/.cache/fenics && rm -r dijitso && cd -
import numpy as np
import sympy as sym
#HTML(X3DOM().html(u))
```

```python
def kappa(u):
    return 1 + u*u

u_exp = Expression('sin(2*pi*x[0]) * cos(2*pi*x[1])', degree=2)
grad_u_exp = Expression(('2*pi*cos(2*pi*x[0])*cos(2*pi*x[1])','-2*pi*sin(2*pi*x[0])*sin(2*pi*x[1])'), degree = 2)

x, y = sym.symbols('x[0], x[1]')  # define x and y in FENICS syntax
u_ex = sym.sin(2*pi*x) * sym.cos(2*pi*y)     # define u
# get f using symbolic differentiation of u
f = - sym.diff(kappa(u_ex)*sym.diff(u_ex, x), x) - sym.diff(kappa(u_ex)*sym.diff(u_ex, y), y)  
f = sym.simplify(f)

u_code = sym.printing.ccode(u_ex)  # generate code for u in C syntax
f_code = sym.printing.ccode(f)  # generate code for f in C syntax
print('u =', u_code)
print('f =', f_code)
```

```python
def error_calc(h, p):
    #mesh = UnitSquareMesh(h, h)   # generate mesh
    mesh = UnitQuadMesh(mpi_comm_world(), h, h)
    V = FunctionSpace(mesh, 'Lagrange', p)

    # Define boundary condition
    def boundary(x, on_boundary):
        return on_boundary

    bc = DirichletBC(V, u_exp, boundary)

    # Define Variational problem
    u = Function(V)
    v = TestFunction(V)
    fa = Expression(f_code, degree=2)
    F = inner(kappa(u)*grad(u), grad(v))*dx - fa*v*dx
    solve(F == 0, u, bc)
    error = (u_exp - u)**2*dx
    error1 = (grad_u_exp - grad(u))**2*dx + (u_exp - u)**2*dx
    l2 = sqrt(abs(assemble(error)))
    h1 = sqrt(abs(assemble(error1)))

    return l2, h1
```

```python
h_list = np.array([5, 10, 20, 40, 60, 80, 100, 125, 150])
p_list = [1, 2, 3]

L2_err = np.zeros((len(p_list), len(h_list)))
H1_err = np.zeros((len(p_list), len(h_list)))

pidx = 0
for pn in p_list:
    hidx = 0
    for hn in h_list:
        L2_err[pidx, hidx], H1_err[pidx, hidx] = error_calc(hn, pn)
        hidx += 1
    pidx += 1
```

```python
plt.style.use('ggplot')
plt.loglog(h_list, L2_err[0, :], 'o-', label='p1')
plt.loglog(h_list, L2_err[1, :], 'o-', label='p2')
plt.loglog(h_list, L2_err[2, :], 'o-', label='p3')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
plt.xlabel("No. elements")
plt.ylabel("L2 error")
plt.show()
```

```python
plt.loglog(h_list, H1_err[0, :], 'o-', label='p1')
plt.loglog(h_list, H1_err[1, :], 'o-', label='p2')
plt.loglog(h_list, H1_err[2, :], 'o-', label='p3')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
plt.xlabel("No. elements")
plt.ylabel("H1 error")
plt.show()
```
